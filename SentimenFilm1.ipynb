```python
#untuk mengumpulkan data tweet sebanyak 100 dengan kata kunci judul film, dan berbahasa indonesia
data_tweet = api.search(q="Avengers Endgame", lang="en", count=500)
```

```python
hasilData = []
#Cleansing data menggunakan library re dengan menghilangkan simbol
for tweet in data_tweet:
    bagian_tweet = {}
    bagian_tweet["Tanggal"] = tweet.created_at
    bagian_tweet["Pengguna"] = tweet.user.screen_name
    bagian_tweet["Tweet"] = tweet.text
    dataClean = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet.text).split())

    analisisData = TextBlob(dataClean)

    if analisisData.sentiment.polarity > 0.0:
        bagian_tweet["Sentimen"] = "Positif"
    elif analisisData.sentiment.polarity == 0.0:
        bagian_tweet["Sentimen"] = "Netral"
    else :
        bagian_tweet["Sentimen"] = "Negatif"

    if tweet.retweet_count > 0: 
        if bagian_tweet not in hasilData:
            hasilData.append(bagian_tweet)
    else:
        hasilData.append(bagian_tweet)
```

```python
tweetPositif = [t for t in hasilData if t["Sentimen"]=="Positif"]
tweetNetral = [t for t in hasilData if t["Sentimen"]=="Netral"]
tweetNegatif = [t for t in hasilData if t["Sentimen"]=="Negatif"]

tweetPositif
```
